```python
import requests
from bs4 import BeautifulSoup
```

```python
def flatten(l):
    return [item for sublist in l for item in sublist]
```

```python
url = 'https://keithgalli.github.io/web-scraping/webpage.html'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
refs1 = soup.select('.socials a')
social_refs = [ref['href'] for ref in refs1]
```

```python
social_refs
```

```python
li_list = soup.select('ul li')
refs2 = []
for li in li_list:
    if li.select('a'):
        refs2.append(li.select('a'))
```

```python
refs2 = [ref['href'] for ref in flatten(refs2)]
```

```python
refs2[1:5]
```
