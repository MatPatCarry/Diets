```python
import numpy as np
```

```python
N = [i for i in range(1, 14)]
ps = np.linspace(0.2, 0.8, 4)
```

```python
ks = np.arange(10, 150, 10)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
options = {
    'node_color': 'red',
    'node_size': 1000,
    'width': 2,
    'edgecolors': 'black'
}
```

```python
fig, ax = plt.subplots(14, 4, figsize=(30, 90))

for row_index, k in enumerate(ks):
    for col_index, p in enumerate(ps):

        w1, w2 = 0, 0
        edges = []

        for _ in range(k):
            w1, w2 = np.random.randint(1, 14, 2)

            while w1 == w2:
                w1, w2 = np.random.randint(1, 14, 2)
            p_random = np.random.uniform()
            if p_random < p:
                edges.append((w1, w2))

        G = nx.Graph()
        G.add_nodes_from(N)
        G.add_edges_from(edges)

        czy_spojny = nx.is_connected(G)
        czy_spojny_label = 'Graf jest spojny' if czy_spojny else 'Graf nie jest spojny'
        
        pos = nx.spring_layout(G)
        ax[row_index, col_index].set_title(f"Ile prb: {k}, jakie prawdopodobiestwo: {round(p, 2)}\n{czy_spojny_label}", fontsize=15)
        nx.draw(G, pos, with_labels=True, font_weight='bold', ax=ax[row_index, col_index], **options)
        
        
plt.show()
```

```python
fig, ax = plt.subplots(5, 3, figsize=(10, 10))
row_index = 0
for col_index, k in enumerate(ks):

    spojny_list = []

    for index, p in enumerate(ps2):

        w1, w2 = 0, 0
        edges = []

        for _ in range(k):
            w1, w2 = np.random.randint(1, 14, 2)

            while w1 == w2:
                w1, w2 = np.random.randint(1, 14, 2)
            p_random = np.random.uniform()
            if p_random < p:
                edges.append((w1, w2))

        G = nx.Graph()
        G.add_nodes_from(N)
        G.add_edges_from(edges)

        czy_spojny = nx.is_connected(G)
        spojny_list.append(1 if czy_spojny else 0)

    ax[row_index, col_index % 3].plot(ps2, spojny_list)
    ax[row_index, col_index % 3].set_title(f'Dla {k} prb')
    if (col_index + 1) % 3 == 0:
        row_index += 1
    
plt.show()
```
