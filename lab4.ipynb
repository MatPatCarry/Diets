```python
with open('duzo_tekstu.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
```

```python
lines = [line.strip() for line in lines]
```

```python
one_str = "".join(lines)
```

```python
one_str
```

```python
splitted = one_str.split()
```

```python
len(splitted)
```

```python
cleared = [word.strip(",;.''-„”") for word in splitted]
```

```python
from collections import Counter
```

```python
counter = Counter(cleared)
```

```python
counter['i']
```

```python
counter.most_common(50)
```

```python
counter_dict_manual = {}
for word in cleared:
    if not word in counter_dict_manual:
        counter_dict_manual[word] = 1
    else:
        counter_dict_manual[word] += 1
```

```python
counter_dict_manual['i']
```

```python
counter_dict_manual.get('i')
```

```python
counter_dict_manual.get('xdddd', 0)
```

```python
counter_dict_manual.get('fff', 4)
```

```python
counter_dict_manual.setdefault('aaaaaaa', -1)
```
